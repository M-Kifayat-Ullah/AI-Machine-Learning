```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

# Generating dataset

```python
data=make_classification(n_samples=5,n_features=2,n_informative=2,n_redundant=0,n_classes=2)
```

```python
data
```

# Tuple Unpacking

## Extracting Feature

```python
x=data[0]
```

```python
y=data[1]
```

```python
x
```

```python
y
```

# Performing Test Splitng

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# performing Min_Max scalling ( range(0,1) for data Normolization)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
sc.fit(x_train)
```

```python
x_train_scalled=sc.transform(x_train)
```

```python
x_train_scalled
```

# Performing test scalling

```python
x_test_scalled=sc.transform(x_test)
```

```python
x_test_scalled
```
